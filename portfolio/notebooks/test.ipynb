```python
import os  
import plotly.io as pio


script_dir = os.getcwd()  # get the current working directory
figures_dir = os.path.join(script_dir, '../reports/figures')
# Define the paths to your saved JSON files
fig1_path = os.path.join(figures_dir, 'plot_cumulative_returns.json')
fig2_path = os.path.join(figures_dir, 'plot_composition.json')
      # Read the figures from the JSON files
with open(fig1_path, 'r') as file:
        fig1 = pio.from_json(file.read())

with open(fig2_path, 'r') as file:
        fig2 = pio.from_json(file.read())


fig1.show()
fig2.show()
```
